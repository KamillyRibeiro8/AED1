```python
import random

numero_secreto = random.randrange(1, 101)
tentativas = 0
limite = 5

print("O número secreto foi gerado! Tente adivinhar.")
print("Tente adivinhar em até 5 tentativas.")

while True :
    if tentativas >= limite:
        print("Você chegou ao limite de tentativas")
        print("FIM DE JOGO! Você perdeu.")
        break
    chute_usuario = int(input("Qual o seu palpite? "))
    if chute_usuario == numero_secreto:
        print("PARABÉNS! Você acertou!")
        break

    elif chute_usuario > numero_secreto:
        print("O seu palpite foi maior que o número secreto.")
    else:
        print("O seu palpite foi menor que o número secreto.")
    tentativas = tentativas + 1
```
